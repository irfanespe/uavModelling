## Data processing

```python
df_take = df[['mavlink_system_time_t.time_boot_ms', 'mavlink_global_position_int_t.lat',
    'mavlink_global_position_int_t.lon', 'mavlink_global_position_int_t.alt', 'mavlink_global_position_int_t.vx',
    'mavlink_global_position_int_t.vy', 'mavlink_global_position_int_t.vz', 'mavlink_wind_t.speed',
    'mavlink_wind_t.speed_z', 'mavlink_wind_t.direction', 'mavlink_vfr_hud_t.airspeed', 'mavlink_vfr_hud_t.groundspeed',
    'mavlink_vfr_hud_t.throttle']]
```

```python
df_take['mavlink_global_position_int_t.alt'] = df_take['mavlink_global_position_int_t.alt'] - 1000000
```

```python
df_take['mavlink_global_position_int_t.alt'] = df_take['mavlink_global_position_int_t.alt'].astype(str).str[:-2] + '.' \
                                               + df_take['mavlink_global_position_int_t.alt'].astype(str).str[-2:]
```

```python
df_take.head(10)
```

```python
test = df_take['mavlink_system_time_t.time_boot_ms'][1:].reset_index().drop(columns = ['index'])
```

```python
lat_next = df_take['mavlink_global_position_int_t.lat'][1:].reset_index().drop(columns = ['index'])
lon_next = df_take['mavlink_global_position_int_t.lon'][1:].reset_index().drop(columns = ['index'])
```

```python
df_take.drop(df.tail(1).index,inplace=True)
```

```python
df_take['mavlink_system_time_t.time_boot_ms_next'] = test
df_take['mavlink_global_position_int_t.lat_next'] = lat_next
df_take['mavlink_global_position_int_t.lon_next'] = lon_next
```

```python
df_take['period'] = df_take['mavlink_system_time_t.time_boot_ms_next'] - df_take['mavlink_system_time_t.time_boot_ms'] 
```

```python
df_take.drop(columns =['mavlink_system_time_t.time_boot_ms', 'mavlink_system_time_t.time_boot_ms_next',
                       'mavlink_wind_t.speed_z'], inplace = True)
```

```python
df_take.columns = ['lat', 'long', 'altitude', 'vx', 'vy', 'vz', 'wind_speed', 'direction', 'airspeed', 'ground_speed',
                   'throttle', 'lat_next', 'lon_next', 'period']
```

```python
df_take = df_take[['period', 'lat', 'long', 'altitude', 'vx', 'vy', 'vz', 'wind_speed', 'direction', 'airspeed', 'ground_speed',
                   'throttle', 'lat_next', 'lon_next']]
```
