```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
```

```python
# read dataset
df = pd.read_csv('2019-09-12 10-14-27_sel_params.csv')
df.head()
```

```python
# erase space in each column name
df.columns = [w.strip() for w in df.columns]
```

## Data Assesment

```python
df.info()
```

```python
df_cp = df.copy()
```

```python
# plt.plot(range(0, df_cp.shape[0]) ,df_cp['mavlink_nav_controller_output_t.nav_roll'])
plt.plot(range(0, df_cp.shape[0]) ,df_cp['mavlink_attitude_t.roll'])
plt.legend()
plt.show()
```

```python
# drop unnecessary parameter
drop_col = ['mavlink_ahrs3_t.roll', 'mavlink_ahrs3_t.pitch', 'mavlink_ahrs3_t.yaw',
            'mavlink_ahrs3_t.altitude', 'mavlink_ahrs3_t.lat', 'mavlink_ahrs3_t.lng',
            'mavlink_attitude_t.time_boot_ms', 'mavlink_global_position_int_t.time_boot_ms',
            'mavlink_gps_rtk_t.time_last_baseline_ms', 'mavlink_gps_rtk_t.baseline_a_mm',
            'mavlink_gps_rtk_t.baseline_b_mm', 'mavlink_gps_rtk_t.baseline_c_mm', 'mavlink_gps_rtk_t.tow',
            'mavlink_gps_rtk_t.accuracy', 'mavlink_gps_rtk_t.rtk_rate', 'mavlink_servo_output_raw_t.time_usec',
            'mavlink_servo_output_raw_t.port', 'mavlink_vfr_hud_t.alt', 'mavlink_gps_raw_int_t.fix_type']

df_sel = df_cp.drop(columns = drop_col)
df_sel.head(20)
```

```python
# add comma in lattitude and longitude
df_sel['mavlink_global_position_int_t.lon'] = df_sel['mavlink_global_position_int_t.lon'].astype(str).str[:-7] + '.' \
                                              + df_sel['mavlink_global_position_int_t.lon'].astype(str).str[-7:]
df_sel['mavlink_global_position_int_t.lat'] = df_sel['mavlink_global_position_int_t.lat'].astype(str).str[:-7] + '.' \
                                              + df_sel['mavlink_global_position_int_t.lat'].astype(str).str[-7:]
```
